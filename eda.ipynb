```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


%matplotlib inline 
```

```python
cd=pd.read_csv('client_data.csv')
cd.head()
```

```python
cd.isnull().sum()# null values are present in the dataset but labled MISSING for that we will replace it with NaN by using replace function
```

```python
cd=cd.replace('MISSING',np.nan)
```

```python
cd.head()
```

```python
print("Missing values percentage for each coulumn:\n",cd.isnull().sum()/len(cd)*100)
```

```python
#now check unique values in each column
print("Unique values in each column:\n",cd.nunique())
```

```python
prd=pd.read_csv('price_data.csv')
prd.head()
```

```python
prd=prd.replace('MISSING',np.nan)
print("Missing values percentage for each coulumn:\n",prd.isnull().sum()/len(prd)*100)
```

```python
#merge both the datasets using common id
data=pd.merge(cd,prd,how="left",on='id')
data.head()
```

```python
data.info()
```

```python
#converting date into datetime64 'date_activ', 'date_end','date_modif_prod','date_renewal' , price_date 
data['date_activ']=pd.to_datetime(data['date_activ'])
data['date_end']=pd.to_datetime(data['date_end'])
data['date_modif_prod']=pd.to_datetime(data['date_modif_prod'])
data['date_renewal']=pd.to_datetime(data['date_renewal'])
data['price_date']=pd.to_datetime(data['price_date'])
data.info()
```

```python
#check if 'date_activ', 'date_end','date_modif_prod','date_renewal' , price_date are null or not
print("Missing values percentage for each coulumn:\n",data.isnull().sum()/len(data)*100)
```

```python
#channel sales has 25 % missing values so we will mod impute it with mode
data['channel_sales']=data['channel_sales'].fillna(data['channel_sales'].mode()[0])
data['channel_sales'].isnull().sum()
```

```python
#unique values in 'channel_sales','has_gas', 'origin_up'
print("Unique values in each column:\n",data[['channel_sales','has_gas', 'origin_up']].nunique())
```

```python
#changing categorical values to numerical values
import  sklearn
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['channel_sales']=le.fit_transform(data['channel_sales'])
data['has_gas']=le.fit_transform(data['has_gas'])
data['origin_up']=le.fit_transform(data['origin_up'])
data[['channel_sales','has_gas', 'origin_up']].head()

```

```python
#moving churn column to the end
data['churn']=data['churn'].astype('int')
churn=data['churn']
data.drop(labels=['churn'],axis=1,inplace=True)
data['churn']=churn
```

```python
#heat map with id removed
plt.figure(figsize=(40,20))
#adding mask to hide upper triangle
mask = np.triu(data.drop('id',axis=1).corr())
sns.heatmap(data.drop('id',axis=1).corr(),annot=True, cmap='BrBG',mask=mask)
```
